```python
import pandas as pd
credit = pd.read_excel('/content/Clientes_Renamed.xlsx')

credit_renamed = credit.rename(columns = {'Primeira_Coluna':'valor_credito',
                                          'SEX':'sexo',
                                          'EDUCATION':'nivel_escolaridade',
                                          'MARRIAGE':'estado_civil',
                                          'AGE':'idade',
                                          'PAY_0':'status_pag_set_2005',
                                          'PAY_2':'status_pag_ago_2005',
                                          'PAY_3':'status_pag_jul_2005',
                                          'PAY_4':'status_pag_jun_2005',
                                          'PAY_5':'status_pag_mai_2005',
                                          'PAY_6':'status_pag_abr_2005',
                                          'BILL_AMT1':'extrato_set_2005',
                                          'BILL_AMT2':'extrato_ago_2005',
                                          'BILL_AMT3':'extrato_jul_2005',
                                          'BILL_AMT4':'extrato_jun_2005',
                                          'BILL_AMT5':'extrato_mai_2005',
                                          'BILL_AMT6':'extrato_abr_2005',
                                          'PAY_AMT1':'valor_pag_set_2005',
                                          'PAY_AMT2':'valor_pag_ago_2005',
                                          'PAY_AMT3':'valor_pag_jul_2005',
                                          'PAY_AMT4':'valor_pag_jun_2005',
                                          'PAY_AMT5':'valor_pag_mai_2005',
                                          'PAY_AMT6':'valor_pag_abr_2005'
})

credit = credit_renamed
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import pandas as pg

```

```python
credit_group = credit.groupby(['nivel_escolaridade']).mean().reset_index()
```

```python
credit_group
```

```python
plt.figure(figsize = (10,8))

plt.xlabel('Nível de Escolaridade')
plt.ylabel('Media de valor de Crédito Concedido')
plt.title('Valor de Credito x Escolaridade')

plt.bar(credit_group['nivel_escolaridade'], credit_group['valor_credito'], color = 'red');
```

```python
fig, axes = plt.subplots(2,2, figsize = (18,10))

sns.barplot(data=credit_group, x = 'nivel_escolaridade', y = 'extrato_jun_2005', ax = axes[0,0]);
sns.barplot(data=credit_group, x = 'nivel_escolaridade', y = 'extrato_jul_2005', ax = axes[0,1]);
sns.barplot(data=credit_group, x = 'nivel_escolaridade', y = 'extrato_ago_2005', ax = axes[1,0]);
sns.barplot(data=credit_group, x = 'nivel_escolaridade', y = 'extrato_set_2005', ax = axes[1,1]);
```

```python
fig = go.Figure()


fig.add_trace(go.Bar(x = credit_group['nivel_escolaridade'], y=credit_group['valor_pag_abr_2005'], marker_color = '#0000FF', name = 'Abril'))
fig.add_trace(go.Bar(x = credit_group['nivel_escolaridade'], y=credit_group['valor_pag_mai_2005'], marker_color = '#FF0000', name = 'Maio'))
fig.add_trace(go.Bar(x = credit_group['nivel_escolaridade'], y=credit_group['valor_pag_jun_2005'], marker_color = '#FF00FF', name = 'Junho  '))

fig.update_layout(yaxis = dict(title="Valor Pago em 2005", titlefont_size = 20, tickfont_size = 12),

                  xaxis = dict(title="Escolaridade", titlefont_size = 20, tickfont_size = 12))
```

```python
fig.write_html('valorxescolaridade.html')
```
