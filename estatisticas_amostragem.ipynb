```python
# amostragem sistemática
# criar uma função para pegar uma amostra sistematica
# np.arange(X, len(df) , step = step) - primeiro termo x, tamanho do dataframe, incremento

def amostragem_sistematica(df, step):
  indexes = np.arange(1, len(df) , step = step)
  amostra_sistematica = df.iloc[indexes]
  return amostra_sistematica

amostra_sistematica = amostragem_sistematica(df, 3)

display(amostra_sistematica)
```

```python
# amostragem clusterizada - criar cluster aleatórios
df2 = df
def amostra_cluster(df, n_por_cluster, n_cluster):

# n = tamanho do df
  n = len(df)
# k = numero de divisões por clusters - vai pegar o tamanho total do dataframe e dividir pelo número de elementos que queremos por cluster
  k = int(n/n_por_cluster)

# inicia a variavel vazia
  data = None



  for i in range(k):

    # pega n_por_cluster elemento aleatórios do dataframe
    sample_i = df.sample(n_por_cluster)
    # cria uma coluna CLUSTER no dataframe - função repeat(a,b) do numpy repete o termo a, b vezes
    # neste caso, ele vai colocar o indice i em na coluna cluster em todos os elementos aleatórios escolhidos na função anterior
    sample_i['cluster'] = np.repeat(i, len(sample_i))
    
    #df = df.drop(index = sample_i.index)
    # concatena as variáveis data e sample_i
    data = pd.concat([data, sample_i], axis = 0)

  # retorna um array escolhendo números aleatórios entre 0 e k e o tamanho é o número de clusters
  escolha_cluster_aleatorio = np.random.randint(0, k, size = n_cluster)

  # pega os clusters aleatórios selecionados na função anterior
  sample = data[data.cluster.isin(escolha_cluster_aleatorio)]

  return sample



sample = amostra_cluster(df = df, n_por_cluster = 30, n_cluster = 2)

#print('df fora do for: ', df)
sample
```
