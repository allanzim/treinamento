```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
dataframe = pd.read_excel('/content/Clientes_Renamed.xlsx')
```

```python
#Plotando histograma:

plt.figure(figsize=(30,10))
sns.set_theme(style="darkgrid")

sns.histplot(data=dataframe, x='idade', hue='sexo', kde=True, palette="dark:#5A9_r");
plt.legend(labels=['Masculino', 'Feminino']);
```

```python
sns.displot(data=dataframe, x='idade', row='sexo', binwidth=3, height=3, facet_kws=dict(margin_titles=True), palette="dark:#5A9_r")
```
