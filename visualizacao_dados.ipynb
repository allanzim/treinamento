```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np

credit = pd.read_csv('/content/credit_card_clients_2.csv')

credit = credit.rename(columns = {'LIMIT_BAL':'valor_credito',
                                          'SEX':'sexo',
                                          'EDUCATION':'nivel_escolaridade',
                                          'MARRIAGE':'estado_civil',
                                          'AGE':'idade',
                                          'PAY_0':'status_pag_set_2005',
                                          'PAY_2':'status_pag_ago_2005',
                                          'PAY_3':'status_pag_jul_2005',
                                          'PAY_4':'status_pag_jun_2005',
                                          'PAY_5':'status_pag_mai_2005',
                                          'PAY_6':'status_pag_abr_2005',
                                          'BILL_AMT1':'extrato_set_2005',
                                          'BILL_AMT2':'extrato_ago_2005',
                                          'BILL_AMT3':'extrato_jul_2005',
                                          'BILL_AMT4':'extrato_jun_2005',
                                          'BILL_AMT5':'extrato_mai_2005',
                                          'BILL_AMT6':'extrato_abr_2005',
                                          'PAY_AMT1':'valor_pag_set_2005',
                                          'PAY_AMT2':'valor_pag_ago_2005',
                                          'PAY_AMT3':'valor_pag_jul_2005',
                                          'PAY_AMT4':'valor_pag_jun_2005',
                                          'PAY_AMT5':'valor_pag_mai_2005',
                                          'PAY_AMT6':'valor_pag_abr_2005' 
                                          })
```

```python
credit.notnull()
```

```python
credit
```

```python
credit2 = credit.copy()
credit2
```

```python
credit2.drop('ID', axis=1, inplace=True)
```

```python
credit2.dropna(axis=0, how='all', inplace=True)

# axis: Se for 0 serão ‘dropadas’ as linhas que contem os dados faltantes, caso 1 remove a coluna inteira que tem dados faltantes;

# how: Determina se a coluna vai ser removida do DataFrame, se passarmos ‘any’ como valor do parâmetro se tiver apenas um valor
# faltante que seja a coluna é removida, se passarmos ‘all’ precisamos que todos os valores sejam faltantes/nulos para que a coluna seja removida;

# thresh: com um inteiro, podemos definir o mínimo de colunas de NaN aceitáveis para começar a dropar, ex: caso tenha thresh = 50, precisamos
# que a coluna tenha mais 50 de valorers faltantes para o método ser aplicado;

# subset: se estamos removendo linhas, podemos inserir um array de colunas onde o metodo será aplicado;

# inplace: caso seja True aplica as alterações no dataset de forma automática;
```

```python
credit3 = credit2.drop_duplicates()

#excluir linhas com determinada condição
indice = credit3[credit3['sexo'] > 2].index
credit3.drop(indice, inplace = True)

#salvando arquivo para excel
credit3.to_excel('Clientes_Renamed.xlsx')
```

```python
credit_group = credit3.groupby(['sexo']).mean().reset_index()

credit_group
```

```python
#divide um intervalo de 0 a 2 em 100 intervalos iguais

t = np.linspace(0, 2, 100)


f1 = np.sin(np.pi * t)
f2 = np.sin(2 * np.pi * t)

plt.figure(figsize=(10,8))
plt.plot(t,f1, linewidth = 2, label = 'sin(pi*t)', color = 'yellow')
plt.plot(t,f2, linewidth = 5, label = 'sin(2*pi*t)', color='blue');

plt.legend();

plt.grid(b=True, color='lightgrey', linestyle='--')
plt.title('Funções');
```
